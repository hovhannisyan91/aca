```python
def add_numbers(a: float, b: float) -> float:
    """
    Add two numeric values and return their sum.

    This function takes two numbers, adds them together, and returns the result.
    Although the parameters are annotated as floats, integers can also be passed
    because they are compatible with numeric addition in Python.

    Parameters
    ----------
    a : float
        The first number to add.
    b : float
        The second number to add.

    Returns
    -------
    float
        The sum of `a` and `b`.

    Examples
    --------
    >>> add_numbers(10.5, 4.5)
    15.0

    >>> add_numbers(2, 3)
    5
    """

    c = a + b
   
    return c
```

### CSV Dowloader 2

```python
def csv_dowloader(csv_files:list, source:str, destination: str, type:str = 'csv'):
    for i in csv_files:
        print(f"trying to read: {i}")
        print(f"source: {source}")
        print(f"destination: {destination}")

        df = pd.read_csv(f"{join(source,i)}.csv")
        print(f"{i} shape: {df.shape}")
        
        file_name = join(destination,f"{i}.{type}")
        print(file_name)
        if type == 'xlsx':
            df.to_excel(file_name,index= False)
        elif type == 'parquet':
            df.to_parquet(file_name, index= False)
        else:
            df.to_csv(file_name,index= False)
        print(10*"=")
    
    return 'Okay'
```

```python
URL = "https://raw.githubusercontent.com/hovhannisyan91/sql-analytics-portfolio/refs/heads/main/data/public_schema"
# DESTINATION = '/Users/karen/Desktop/aca/lab/sql_module/data/public_schema'
DESTINATION = '../data/public'
csv_dowloader(csv_files= csv_files,source=URL,destination=DESTINATION,type='xlsx')
```

```python
def score_label(score):
    if score >= 90:
        return "Excellent"
    elif score >= 70:
        return "Good"
    elif score >= 50:
        return "Pass"
    else:
        return "Fail"

score_label(85)
```

```python
import pandas as pd

df_scores = pd.DataFrame({
    "student": ["Anna", "Ben", "Chris", "Diana", "Eva"],
    "score": [95, 78, 61, 43, 88]
})

df_scores
```

```python
df_scores['label_lc'] = [score_label(i) for i in df_scores['score']]
df_scores
```

```python
df_products_dummy = pd.DataFrame({
    "product": ["Milk", "Bread", "Juice", "Cheese", "Steak", "Apples"],
    "price": [2.5, 1.8, 6.2, 12.0, 24.5, 4.2]
})

df_products_dummy
```

```python

def classify_price(price:float, low:int=5, high:int=15):
    if price <= low:
        return "Low-range product"
    elif price <= high:
        return "Mid-range product"
    else:
        return "High-range product"

classify_price(9)
```

```python
df_products_dummy['classify_price'] = df_products_dummy['price'].apply(classify_price)
```

```python
def income_band(income, low=30000, high=70000):
    if income < low:
        return "Low income"
    elif income < high:
        return "Middle income"
    else:
        return "High income"
```

```python
df_customers_dummy = pd.DataFrame({
    "customer": ["A", "B", "C", "D", "E"],
    "age": [22, 35, 47, 63, 29],
    "income": [18000, 42000, 72000, 95000, 25000]
})

df_customers_dummy
```

```python

df_customers_dummy["income_group"] = df_customers_dummy["income"].apply(income_band)
df_customers_dummy
```

```python

df_customers_dummy["income_group"] = (df_customers_dummy["income"]
                                      .apply(income_band,
                                                        low = 15000,
                                                        high = 50000))
df_customers_dummy
```

```python
df_customers_dummy[['income','income_group']].groupby('income_group').mean()
```
