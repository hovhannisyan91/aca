```python
import pandas as pd
import numpy as np
import plotly.express as px

np.random.seed(42)


months = pd.date_range("2024-01-01", periods=6, freq="MS")
regions = ["North", "South", "East"]
products = ["A", "B"]

rows = []

for month in months:
    for region in regions:
        for product in products:
            sales = np.random.randint(80, 220)
            customers = np.random.randint(20, 90)
            units = np.random.randint(10, 70)
            
            rows.append([month, region, product, sales, customers, units])

df_plotly = pd.DataFrame(
    rows,
    columns=["month", "region", "product", "sales", "customers", "units"]
)

df_plotly["month_name"] = df_plotly["month"].dt.strftime("%b")
df_plotly.head()
```

```python
from utils_ploty import *
```

```python
px.defaults.template = "plotly_white"
```

## Bar Plot `Sales by Region`

[Execution order](https://hovhannisyan91.github.io/aca/materials/sql/slides/session03.html#/execution-order)

```python
region_sales = df_plotly.groupby('region', as_index=False)['sales'].sum()
region_sales
```

```python
fig = px.bar(
    data_frame  = region_sales,
    x = 'region',
    y = 'sales',
    title = "Total Sales by Region",
    text = 'sales'   
)

fig.update_traces(
    textposition = 'outside'
)

fig.update_layout(
    xaxis_title = '',
    yaxis_title = 'Sales'
)
fig.show()
```

## Line Chart highlight

```python
monthly_sales
```

```python
min_sales_idx = monthly_sales['sales'].idxmin()

min_monthly_sales = monthly_sales.loc[min_sales_idx]
x_min = min_monthly_sales[['month']]
y_min = min_monthly_sales[['sales']]
```

```python
max_sales_idx = monthly_sales['sales'].idxmax()

max_monthly_sales = monthly_sales.loc[max_sales_idx]
x_max = max_monthly_sales[['month']]
y_max = max_monthly_sales[['sales']]
```

```python
monthly_sales_region
```

```python
fig = px.line(
    data_frame = monthly_sales,
    x = 'month',
    y = 'sales'
    # color = 'region',
    # markers = True
)

fig.add_scatter(x = x_max, 
                y = y_max,
                mode = 'markers',
                name = 'Maximum',
                marker=dict(size=14, color="#ABEAA5"),
                )

fig.add_scatter(x = x_min, 
                y = y_min,
                mode = 'markers',
                name = 'Minimum',
                marker=dict(size=14, color="#B22E24"),
                )

fig.update_layout(
    # x
    # y
    # legend_title  = 'Region'
)

fig.show()
```

```python
df_hist = df_plotly.copy()

df_hist["highlight"] = np.where(
    df_hist["region"] == top_region,
    "Highest Sales Region",
    "Other Regions"
)

df_hist.head()
```

```python
fig = px.histogram(
    df_hist,
    x="sales",
    color="highlight",
    barmode="overlay",
    opacity=0.5,
    title="Distribution with Highlighted Largest Category",
    color_discrete_map={
        "Highest Sales Region": "#3B6EAD",
        "Other Regions": "#D9D9D9"
    }
)

fig.update_layout(
    xaxis_title="Sales",
    yaxis_title="Count",
    legend_title=""
)

fig.show()
```
