```python
import pandas as pd

df_dates = pd.DataFrame({
    "customer_id": [101, 102, 103, 104, 105],
    "signup_date": ["2024-01-15", "2024-02-20", "2024-03-05", "2024-03-18", "2024-04-10"],
    "last_purchase": ["2024-06-01", "2024-06-15", "2024-07-01", "2024-07-08", "2024-07-12"]
})

df_dates.to_csv('test.csv',index=False)
```

```python
df_dates = pd.read_csv('test.csv',
                       parse_dates=['signup_date','last_purchase'])
df_dates.info()
```

```python
df_dates.info()
```

```python
df_dates['new_signup_date'] = [datetime.strptime(i,"%Y-%m-%d") for i in df_dates['signup_date']]
```

```python
df_dates['new_day'] = [i + timedelta(days=1) for i in df_dates['new_signup_date']]
df_dates
```

```python
df_dates['signup_date'] = pd.to_datetime(df_dates['signup_date'])
```

```python
df_dates.drop(columns=df_dates[['new_signup_date_1']],inplace=True)
```

```python
(df_dates["signup_date"].dt.dayofweek >= 5).astype(int)
```

```python
df_dates["is_weekend"] = (df_dates["signup_date"].dt.dayofweek >= 5).astype(int)
df_dates[["signup_date", "signup_dow", "is_weekend"]]
```

```python
df_dates['signup_date_don'] = df_dates['signup_date'].dt.day_name()
df_dates['signup_date_dow'] = df_dates['signup_date'].dt.day_of_week
df_dates['signup_date_dm'] = df_dates['signup_date'].dt.daysinmonth

df_dates
```

```python
pd.concat([df_dates, pd.get_dummies(df_dates['signup_date_don'],drop_first=True)], axis=1)
```

```python
df_dates['last_purchase'] = pd.to_datetime(df_dates['last_purchase'])
```

```python
df_dates['last_purchase_dow']= df_dates['last_purchase'].dt.day_of_week
df_dates

df_dates['last_purchase_doy']= df_dates['last_purchase'].dt.day_of_year
df_dates
```

```python
df_dates['signup_month_sm'] = df_dates['signup_date'].dt.to_period("M").dt.start_time
df_dates['signup_month_em'] = df_dates['signup_date'].dt.to_period("M").dt.
df_dates
```
